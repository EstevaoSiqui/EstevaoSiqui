## Limpeza dos textos

```python

import re
# Removendo caracteres especiais
datatest["resource"] = datatest["resource"].apply(lambda text: re.sub(r"[^\w\s]", "", text))
datatest["text"] = datatest["text"].apply(lambda text: re.sub(r"[^\w\s]", "", text))
datatest["event"] = datatest["event"].apply(lambda text: re.sub(r"[^\w\s]", "", text))

```

```python
#convertendo de maiusculo para minusculo
datatest["resource"] = datatest["resource"].str.lower()
datatest["text"] = datatest["text"].str.lower()
datatest["event"] = datatest["event"].str.lower()
```

```python
import spacy
from spacy.lang.pt import stop_words

nlp = spacy.load("pt_core_news_sm")

# Removendo stop words e reconstruindo o texto
for index, row in datatest.iterrows():
  text = row["text"]

  # Tokenização
  doc = nlp(text)

  # Removendo stop words
  new_text = " ".join([token.text for token in doc if not token.is_stop])

  # Reconstruindo o texto
  datatest.loc[index, "text"] = new_text

# Lemmatization e reconstrução do texto
for index, row in datatest.iterrows():
  text = row["text"]

  # Tokenização
  doc = nlp(text)

  # Lemmatization
  new_text = " ".join([token.lemma_ for token in doc])

  # Reconstruindo o texto
  datatest.loc[index, "text"] = new_text
```
